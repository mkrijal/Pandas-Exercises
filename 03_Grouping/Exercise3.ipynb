### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|',index_col = 'user_id' )
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
groupby_occupation = users['age'].groupby(users['occupation'])

groupby_occupation.mean()

```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
male_users = users[(users.gender == 'M')] # male users dataframe

#total male age for each occupation compare to overall age for each occupation
m = users.groupby([male_users['occupation']]).size() * 100 / users.groupby([users['occupation']]).size()

#sort by values
m.sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
#function min and max
def get_stats(group):
    return {'min': group.min(), 'max': group.max()}

#group by occupation
users['age'].groupby(users['occupation']).apply(get_stats).unstack()
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users['age'].groupby([users['occupation'], users['gender']]).mean().unstack()
```

### Step 8.  For each occupation present the percentage of women and men

```python
g = users.groupby([users['occupation'], users['gender']]).size() * 100 / users.groupby([users['occupation']]).size()
g.unstack()
```
